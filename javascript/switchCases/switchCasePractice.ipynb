<h1>Switch Cases</h1><br>
<p>The switch statement evaluates an expression and executes code as a result of a matching case. The basic syntax is similar to that of an if statement. It will always be written with switch () {}, with parentheses containing the expression to test, and curly brackets containing the potential code to execute.</p><br>
<p>Using switch, we will send a message to the console each day of the week. The program will run in order from top to bottom looking for a match, and once one is found, the break command will halt the switch block from continuing to evaluate statements.</p>

```
const day = new Date().getDay();
//sets the current day of the week to the day variable

switch (day) {
//switch will read top to bottom once it reaches the day of the week, the break statement will halt the switch from continuing. 
    case 0:
        console.log("Sunday");
        break;
        //could be done withough break statements, but using them keeps the program from checking even after it reaches the cooresponding day.
        //it is more proficient this way.
    case 1:
        console.log("Monday");
        break;
    case 2:
        console.log("Tuesday");
        break;
    case 3:
        console.log("Wednesday");
        break;
    case 4:
        console.log("Thursday");
        break;
    case 5:
        console.log("Friday");
        break;
    case 6:
        console.log("Saturday");
        break;
    default:
        console.log("WHat?")
}
```

<h1>Switch Ranges</h1><br>
<h3>There might be an occasion in which you will need to evaluate a range of values in a switch block, as opposed to a single value as in our example above. We can do this by setting our expression to true and doing an operation within each case statement.</h3>
<p>To make this easier to understand, we will use a familiar example. In the conditional statements tutorial, we made a grading app which would take a number score and convert it to a letter grade, with the following requirements.
<ul>
<li>Grade of 90 and above is an A</li>
<li>Grade of 80 to 89 is a B</li>
<li>Grade of 70 to 79 is a C</li>
<li>Grade of 60 to 69 is a D</li>
<li>Grade of 59 or below is an F</li>
</ul>

Now we can write that as a switch statement. Since weâ€™re checking a range, we will perform the operation in each case to check if each expression is evaluating to true then break out of the statement once the requirements for true have been satisfied.</p>

```
//set the grade
const grade = 87;

switch (true) {
    //if grade is 90 or above
    case grade >= 90:
        console.log("A");
        break;
    //if grade is 80 or above
    case grade >= 80:
        console.log("B");
        break;
    //if grade is 70 or above
    case grade >= 70:
        console.log("C");
        break;
    //if grade is 60 or above
    case grade >= 60:
        console.log("D");
        break;
    //anything below 59 is failing
    default:
        console.log("F");
}
```

<h1>Multiple Cases</h1><br>
<p>You may encounter code in which multiple cases should have the same output. In order to accomplish this, you can use more than one case for each block of code.

In order to test this, we are going to make a small application matching the current month to the appropriate season. First, we will use the new Date() method to find a number corresponding to the current month, and apply that to the month variable. The new Date().getMonth() method will output a number from 0 to 11, with 0 being January and 11 being December. At the time of this publication, the month is September, which will correspond to 8.

Our application will output the four seasons with the following specifications for simplicity:
<ul>
<li>Winter: January, February, and March</li>
<li>Spring: April, May, and June</li>
<li>Summer: July, August, and September</li>
<li>Autumn: October, November, and December</li>
</ul>

Below is our code.

```
//get number corresponding to current month, with 0 being Jan and 11 being Dec
const month = new Date().getMonth();

switch (month) {
    //Jan, Feb, Mar
    case 0:
    case 1:
    case 2:
        console.log("Winter");
        break;
    //Apr, May, Jun
    case 3:
    case 4:
    case 5:
        console.log("Spring");
        break;
    //July, Aug, Sept
    case 6:
    case 7:
    case 8:
        console.log("Summer");
        break;
    case 9:
    case 10:
    case 11:
        console.log("Fall");
        break;
    default:
        console.log("Something is wrong...");
}
```
